```python
from helpers import *
torch.manual_seed(7)
np.random.seed(7)
```

```python
class neural_net(torch.nn.Module):
    def __init__(self, input_size, output_size, hidden_units):
        super(neural_net, self).__init__()
        self.fc1 = torch.nn.Linear(input_size, hidden_units)
        self.fc2 = torch.nn.Linear(hidden_units, output_size)
        self.phi = torch.relu
    def forward(self, x):
        x = self.phi(self.fc1(x))
        x = torch.abs(self.fc2(x))
        return x 
```

```python
n_mfcc = 35
max_signal_len = 7000
max_pad_len = 20
lim = 2

#input, target, test_indices = get_data(n_mfcc, max_signal_len, max_pad_len, lim)
input, target, test_indices = augmented_data(n_mfcc, max_signal_len, max_pad_len, lim)
train_input, train_target = input[test_indices == 0], target[test_indices == 0]
test_input, test_target = input[test_indices == 1], target[test_indices == 1]
print(train_input.shape)
print(test_input.shape)

train_input = torch.from_numpy(train_input).type(torch.FloatTensor)
train_classes = torch.from_numpy(train_target).type(torch.LongTensor)
test_input = torch.from_numpy(test_input).type(torch.FloatTensor)
test_classes = torch.from_numpy(test_target).type(torch.LongTensor)
```

```python
indices = torch.randperm(train_input.shape[0])
train_input = train_input[indices]
train_classes = train_classes[indices]

train_input, mean, std = standardize(train_input)
test_input = (test_input - mean) / std

hidden_units = 4
learning_rate = 1
nb_epochs = 200

input_dim = input.shape[1]
output_dim = 10
model = neural_net(input_dim, output_dim, hidden_units)
optimizer = torch.optim.Adadelta(model.parameters())
train_loss_history, test_loss_history, train_acc_history, test_acc_history = train_ce(model, train_input, train_classes, test_input, test_classes, optimizer, nb_epochs, batch_size=40)
```
